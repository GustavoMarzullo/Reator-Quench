```python
from Reator.reator import Reagente, Leito, Reator
import matplotlib.pyplot as plt
import numpy as np
```

```python
F=6500
reagente = Reagente(
    FN2=0.21825*F,
    FH2=0.65475*F,
    FNH3=0.05*F,
    P=225,
    T=410)

L1 = Leito(
    L = 5)

L2 = Leito(
    L = 10,
    T = 410)

L3 = Leito(
    L = 10,
    T = 410)

reator = Reator(
    Leitos=[L1, L2, L3],
    Reagente=reagente,
    Y = [0.7, 0.2, 0.1])

resultado = reator.solve()
resultado.plot()
```

```python
resultado.x()
```
