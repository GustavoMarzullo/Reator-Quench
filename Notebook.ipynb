```python
from Reator.reator import Reagente, Leito, Reator
```

```python
F=6500
reagente = Reagente(
    FN2=0.21825*F,
    FH2=0.65475*F,
    FNH3=0.05*F,
    P=180,
    T=420)
```

```python
L1 = Leito(
    L = 2)

L2 = Leito(
    L = 5,
    T = 400)

L3 = Leito(
    L = 15,
    T = 400)

reator = Reator(
    Leitos=[L1, L2, L3],
    Reagente=reagente,
    Y = [0.5, 0.3, 0.2])
```

```python
resultado = reator.solve()
```

```python
resultado.plot()
```
