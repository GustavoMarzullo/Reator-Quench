```python
from Reator.reator import Reagente, Leito, Reator
```

```python
F=6500
reagente = Reagente(
    FN2=0.21825*F,
    FH2=0.65475*F,
    FNH3=0.05*F,
    P=220,
    T=420)
```

```python
L1 = Leito(
    L = 5)

L2 = Leito(
    L = 10,
    T = 400)

L3 = Leito(
    L = 10,
    T = 400)

reator = Reator(
    Leitos=[L1, L2, L3],
    Reagente=reagente,
    Y = [0.6, 0.2, 0.2])
```

```python
resultado = reator.solve()
```

```python
resultado.plot()
```

```python
reator2 = Reator(
    Leitos = [L1],
    Reagente = reagente,
    Y=[1])
P = list(range(155,295))
XN2 = []
for i in P:
    reator2.Reagente.P = i
    resultado = reator2.solve()
    XN2.append(resultado.XN2())
```
